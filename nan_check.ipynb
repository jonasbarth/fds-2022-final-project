# Checking files that have nan
I generated two datasets, one with file names where **all** values are `nan`, and one with filenames where at least **one** but not all values are `nan`.

the script used to generate these datasets is in `scripts/find_nan_image.py`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## All nan
There are 90 images where all values for all 15 channels are `nan`.

```python
all_nan = pd.read_csv("script/out/all_nan_files.csv", index_col=0)
all_nan
```

## Showing an image
Showing the RGB channels of the image

```python
nan_image = np.load(f'data/train/{all_nan.filename.loc[0]}')
plt.imshow(nan_image[:,:,:3])
```
